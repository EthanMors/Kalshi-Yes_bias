```python
summary_df['close_time'] = pd.to_datetime(summary_df['close_time'], utc=True)

# 2. Shift it to our strict Pacific Standard Time (-08:00)
summary_df['close_time'] = summary_df['close_time'].dt.tz_convert('-08:00')

# 3. Extract just the local date! 
summary_df['local_date'] = summary_df['close_time'].dt.date

# Check our work to make sure the local_date matches the date in the ticker!
print(summary_df[['ticker', 'close_time', 'local_date', 'result']].head())
```

```python
master_trades_df = pd.read_csv("all_LA_weather_trade_data_3-2-26.csv")
```

```python
master_trades_df.tail()
```

```python
print("Converting trade timestamps and extracting keys...")

# 1. Convert the raw strings into Pandas datetime objects, forcing them to be recognized as UTC
master_trades_df['created_time'] = pd.to_datetime(master_trades_df['created_time'], utc=True)

# 2. Convert from UTC to our strict NWS timezone (-08:00)
master_trades_df['created_time'] = master_trades_df['created_time'].dt.tz_convert('-08:00')

# 3. Extract the 'local_date' and 'hour' keys
# .dt.date drops the time completely, .dt.hour extracts just the 0-23 hour integer
master_trades_df['local_date'] = master_trades_df['created_time'].dt.date
master_trades_df['hour'] = master_trades_df['created_time'].dt.hour

# Let's check our work!
print(master_trades_df[['created_time', 'local_date', 'hour', 'yes_price']].head())
```

```python
# Squash the individual ticks into hourly chunks
df_hourly_trades = master_trades_df.groupby(['ticker', 'local_date', 'hour']).agg(
    hourly_avg_yes_price=('yes_price', 'mean'),
    hourly_avg_no_price=('no_price', 'mean'),
    hourly_volume=('count', 'sum')
).reset_index()

print("Trade data successfully squashed into hourly rows!")
print(df_hourly_trades.head())
```

```python
from datetime import datetime
import meteostat as ms

lax = ms.Point(33.9416, -118.4085)
START = datetime(2025, 1, 4)
END = datetime(2026,3,3)
stations = ms.stations.nearby(lax,limit=1)
daily_data = ms.daily(stations,START,END)
df_daily_weather = daily_data.fetch()
```

```python
df_daily_weather['temp'] = (df_daily_weather['temp'] * 9/5) + 32
df_daily_weather['tmin'] = (df_daily_weather['tmin'] * 9/5) + 32
df_daily_weather['tmax'] = (df_daily_weather['tmax'] * 9/5) + 32
df_daily_weather
```

```python
hourly_data = ms.hourly(stations,START,END)
df_hourly_weather = hourly_data.fetch()
df_hourly_weather['temp'] = (df_hourly_weather['temp'] * 9/5) + 32
df_hourly_weather
```

```python
df_hourly_clean = df_hourly_weather.reset_index()

# 2. Tell Pandas the time is UTC, then convert it to strict PST (UTC-8)
# We use -08:00 instead of 'America/Los_Angeles' so Daylight Saving doesn't mess it up
df_hourly_clean['time'] = df_hourly_clean['time'].dt.tz_localize('UTC').dt.tz_convert('-08:00')

# 3. Set the corrected time back as the index so Pandas can do the math
df_hourly_clean = df_hourly_clean.set_index('time')
```

```python
print(df_hourly_clean.head())
```

```python
df_weather = df_hourly_clean[["temp","rhum","wdir","wspd","pres","cldc","coco"]].copy()
```

```python
import numpy as np

print("Transforming wind direction into cyclical ML features...")

# 1. Trigonometry requires radians, not degrees. 
# We convert the wdir column to radians first.
df_weather['wdir_rad'] = np.deg2rad(df_weather['wdir'])

# 2. Calculate the Sine and Cosine features
df_weather['wdir_sin'] = np.sin(df_weather['wdir_rad'])
df_weather['wdir_cos'] = np.cos(df_weather['wdir_rad'])

# 3. Drop the raw degrees and the intermediate radian column.
# If you leave the raw 'wdir' column in, the model will still try to use it and get confused!
df_weather = df_weather.drop(columns=['wdir', 'wdir_rad'])

# Check your new, machine-learning-ready features!
```

```python
print(df_weather.head())
```

```python
df_weather = pd.get_dummies(df_weather, columns=['coco'], prefix='condition', dtype=int)

# Let's see your final, locked-and-loaded ML feature matrix!
print(df_weather.head())
```

```python
df_weather = df_weather.reset_index()

# 2. Extract the matching keys using the .dt accessor
df_weather['local_date'] = df_weather['time'].dt.date
df_weather['hour'] = df_weather['time'].dt.hour

# Let's verify the keys are there alongside your engineered features!
print(df_weather[['time', 'local_date', 'hour', 'temp', 'wdir_sin']].head())
```

```python
df_weather.to_csv("weather_only.csv", index = False)
```

```python
df_master_hourly = pd.merge(
    df_hourly_trades,
    df_weather,
    on=['local_date', 'hour'],
    how='inner'
)

# 2. Second Merge: Broadcast the Daily Summary outcome to every single hour
# We only bring over the specific columns we need from df_summary to avoid clutter
df_golden = pd.merge(
    df_master_hourly,
    summary_df[['ticker', 'local_date', 'result']], 
    on=['ticker', 'local_date'],
    how='inner'
)

# 3. Clean up the Target Variable for Machine Learning
# Scikit-Learn cannot read the words "yes" or "no". It needs math.
df_golden['target_win'] = df_golden['result'].map({'yes': 1, 'no': 0})

# 4. Optional: Drop the original text 'result' column now that we have our binary target
df_golden = df_golden.drop(columns=['result'])

# Save it to your hard drive so you never have to run this massive pipeline again!
df_golden.to_csv("kalshi_weather_matrix.csv", index=False)

print(f"Success! The Golden Matrix has {len(df_golden)} rows.")
print(df_golden.head())
```
